```python
y = df.Salary
x = df['Percentage in Grade 10']
```

```python
import statsmodels.api as sm
x = sm.add_constant(x)
print(x.head())
print(y.head())
```

```python
#splitting the data

from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size = .8, random_state=100)
```

```python
#fitting the Model 

model_m = sm.OLS(y_train, x_train).fit()
#result = model_m.fit()
model_m.summary2()
```

```python
#printing the estimated parameters
model_m.params
#beta 0 = 30587.285652(intercept)
#beta 1 = 3560.587383(SLOPE)
```

```python
model_m.predict([1,30])
```

```python
import matplotlib.pyplot as plt

def get_std_values(vals):
  return(vals - vals.mean())/vals.std()

x_axis = get_std_values(model_m.fittedvalues)
y_axis = get_std_values(model_m.resid)

plt.scatter(x_axis, y_axis, c = 'g')
plt.xlabel('standardised predicted values')
plt.ylabel('standardised residual values')
plt.title('Residual plot')
```

```python
from scipy.stats import zscore
df.zscore_salary = zscore(df.Salary)
df[(df.zscore_salary > 3.0) | (df.zscore_salary < -3.0)]



# hence there are no out liers
```

```python
##

import numpy as np

mba_influence = model_m.get_influence()
(c,p) = mba_influence.cooks_distance
plt.stem(np.arange(len(x_train)), np.round(c,3), 'ro--')
plt.xlabel('ROW INDEX')
plt.ylabel('COOKS DISTANCE')
```

```python
"""MAking the predictions on validation set 
 and measuring accuracy - R - squared and RMSE  """
import sklearn as ak
from sklearn.metrics import r2_score, mean_squared_error


y_pred = model_m.predict(x_test)
print('R2 Score = ', np.abs(r2_score(y_test, y_pred)))
print('RMSE = ', np.sqrt(mean_squared_error(y_test, y_pred)))
```
